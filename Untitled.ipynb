```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Retail_Transaction_Dataset.csv')
df.head()
df.describe()
df.info()
df.isnull().sum()
```

```python
sns.histplot(df['TotalAmount'])
plt.title('Total Amount Distribution')
plt.show()
```

```python
num_vars = df.select_dtypes(include={'number'})
sns.heatmap(num_vars)
plt.show()
```

```python
plt.hist(df['DiscountApplied(%)'], bins = 20)
plt.show()
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])

# Extract month from 'TransactionDate'
df['Month'] = df['TransactionDate'].dt.month

# Group by month and sum 'TotalAmount'
spend_by_month = df.groupby('Month')['TotalAmount'].sum()

# Plot the bar chart
plt.bar(spend_by_month.index, spend_by_month.values)
plt.title('Total Spend by Month')
plt.xlabel('Month')
plt.ylabel('Total Spend')
plt.xticks(spend_by_month.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
customer_spend = df.groupby('CustomerID')['TotalAmount'].sum()
```

```python
top_customers = customer_spend.sort_values(ascending = False).head(10)
plt.figure(figsize=(10,8))
plt.bar(top_customers.index.astype(str), top_customers.values)
plt.title('Top customers by spend')
plt.xlabel('Customer ID')
plt.show()
for customer_id in top_customers.index:
    store_location = df.loc[df['CustomerID'] == customer_id, 'StoreLocation'].unique()
    for location in store_location:
        print(f"Customer ID: {customer_id}, Store Location : {location}")
```

```python
bottom_customers_spend = df.groupby('CustomerID')['TotalAmount'].sum()
bottom_customers = bottom_customers_spend.sort_values(ascending = False).tail(10)
plt.figure(figsize=(10,8))
plt.bar(bottom_customers.index.astype(str), bottom_customers.values)
plt.title('Bottom 10 customers by spend')
plt.xlabel('Customer ID')
plt.show() 
for customer_id in bottom_customers.index:
    store_location = df.loc[df['CustomerID'] == customer_id, 'StoreLocation'].unique()
    for location in store_location:
        print(f"Customer ID: {customer_id}, Store Location: {location}")
```

```python
avg_discount = df.groupby('ProductCategory')['DiscountApplied(%)'].mean()
plt.figure(figsize=(8,6))
plt.bar(avg_discount.index.astype(str), avg_discount.values)
plt.title('Average discount by product category')
plt.xlabel('Product Category')
plt.show()
```

```python
# Average discount by month
avg_discount_by_month = df.groupby('Month')['DiscountApplied(%)'].mean()
plt.bar(avg_discount_by_month.index, avg_discount_by_month.values)
plt.title('Average discount by month')
plt.xlabel('Month')
plt.ylabel('Average discount')
plt.xticks(avg_discount_by_month.index, ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
max_discount_by_month = df.groupby('Month')['DiscountApplied(%)'].max()
mean_discount_by_month = df.groupby('Month')['DiscountApplied(%)'].mean()
max_spend_in_month = df.groupby('Month')['TotalAmount'].mean()
monthly_sales = df.groupby('Month')['TotalAmount'].sum()
highest_sales_month = monthly_sales.sort_values(ascending = False).head()
print("Average discount in months: ", mean_discount_by_month)
pd.options.display.float_format = '{:,.2f}'.format
print("Highest sales months: ", highest_sales_month)
```

```python
# Scatter plot of Discount vs. Total Spend
plt.figure(figsize=(10, 6))
plt.scatter(df['DiscountApplied(%)'], df['TotalAmount'])
plt.title('Correlation between Discount and Total Spend')
plt.xlabel('Discount')
plt.ylabel('Total Spend')
plt.grid(True)
plt.show()
```

Inferred from the scatter plot above, Discount and Spend show a negative relationship where an increase in discounts applied show a decrease in the total spend, which is unexpected.

```python
prod = df.groupby('ProductCategory')['TotalAmount'].sum()
plt.figure(figsize=(8,6))
plt.bar(prod.index, prod.values)
plt.title('Product category')
plt.xlabel('Product Category')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='ProductCategory', hue='PaymentMethod', palette = 'Paired')
plt.title('Payment Methods Used for Different Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1), loc = 'upper left')
plt.show()
```

```python
# Check product sales by month to see if there's any patterns with that
prod_monthly_sales = df.groupby(['Month', 'ProductCategory'])['TotalAmount'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data = prod_monthly_sales, x = 'Month', y = 'TotalAmount', hue = 'ProductCategory', marker = 'o')
plt.title('Monthly sales by product Category')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

Systemic sales - sales in January were fairly high. Seen a plummet in Febuary for all categories.Then a rise in all for March, April saw a decline again. May saw a rise. Product wise sales analysis - 

Electronics - Fluctuations in sales after May where they decrease in June and don't have a huge spike for a few months where it remains fairly stagnant with gentle rises in sales in July and then a gentle fall in September. Sudden spike in sales in October after which sales fall in November and December.

Books - Started with the lowest sales in January as compared to other categories. May onwards - Gentle drop in sales in June followed by a spike in July and a small fall but fairly stagnant in August. Big fall in sales in September that carries over to October. November and December both see an increase in sales with December seeing the highest spike in the year. Presumably related to the holidays.

Home Decor - Started with the highest sales in comparison to the other products in January. May onwards - June witnesses a slight decline in sales which is then followed by a sharp spike in July. August sees a massive drop in sales which is then followed by an increase in September. Slight drop in October, then a rise in November and December.

Clothing - May onwards - Slight decline in sales in June followed by a slight increase in July. Spike in sales in August followed by a rapid decline in September and even lower in October. Gentle decrease in November followed by a big spike in December, ending the year with the second highest sales in product categories following books.

The general pattern of sales spike in December which can be presumed to be due to the holidays. The turbulent months range from May to October where product category sales fluctuate quite a bit after systemic rises and falls in the early months of the year followed by similar systemic changes in the last couple of months in the year.
The turbulent months could have something to do with discounted prices where the average discount is 10%. January holds a higher discount rate which justifies the high sales which then fall in Febuary.The discount applied in October is higher than the rates from May to September, however, October witnesses a decrease in most sales apart from a rise in electronics which sees its highest sale spike in that month.

```python
payment_method = df.groupby(['Month', 'PaymentMethod'])['TotalAmount'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data = payment_method, x = 'Month', y = 'TotalAmount', hue = 'PaymentMethod', marker = 'o')
plt.title('Monthly sales by Payment Method')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Payment Method')
plt.tight_layout()
plt.show()
```

Debit cards produce the biggest revenue evidenced by a giant spike in July pulling in about 560000 in revenue. The lowest revenue producing payment method is PayPal producing less than 460000 in revenue in April. Big fluctuations in cash payments with drastic rises and falls making it the most turbulent of payment methods. Credit cards  and debit cards seem like the most consistent forms of payment.
May till October seem to be the most active months for producing revenue with special focus on Debit card payments and cash payments.
